```python
"вероятность получения лида = (p-q)^2, где q - значение от 1 до 4, означающее принадлежность к одному из 4-х квартилей по"
"количеству желаемых лидов, а p - значение от 1 до 4, означающее принадлежность к одному из 4-х квартилей по цене, которую"
"байер готов предложить за лид. Таким образом, между байерами с одинаковой ценой в приоритете те, кому нужно 10,20,50 заявок.
```

```python
p1 = np.percentile(d['lead_average_price'], 25)
p2 = np.percentile(d['lead_average_price'], 50)
p3 = np.percentile(d['lead_average_price'], 75)
p4 = np.percentile(d['lead_average_price'], 100)
```

```python
q1 = np.percentile(d['lead_amount'], 25)
q2 = np.percentile(d['lead_amount'], 50)
q3 = np.percentile(d['lead_amount'], 75)
q4 = np.percentile(d['lead_amount'], 100)
```

```python
d['q'] = 0
d['p'] = 0
d
```

```python
for row in d:
    if int(row['lead_amount']) < q1: row['q'] = 1
    elif int(row['lead_amount']) < q2: row['q'] = 2
    elif int(row['lead_amount']) < q3: row['q'] = 3
    else: row['q'] = 4
    if int(row['lead_average_price']) < p1: row['p'] = 1
    elif int(row['lead_average_price']) < p2: row['p'] = 2
    elif int(row['lead_average_price']) < p3: row['p'] = 3
    else: row['p'] = 4
```

```python
dic={}
for i, row in d.iterrows():
    if row['lead_amount'] < q1: row['q'] = 1
    elif row['lead_amount'] < q2: row['q'] = 2
    elif row['lead_amount'] < q3: row['q'] = 3
    else: row['q'] = 4
    if row['lead_average_price'] < p1: row['p'] = 1
    elif row['lead_average_price'] < p2: row['p'] = 2
    elif row['lead_average_price'] < p3: row['p'] = 3
    else: row['p'] = 4    
    dic[i] = ((row['p']-row['q'])**2)
```

```python
k = list(dic.keys())
v = list(dic.values())
f = pd.DataFrame(data=[k,v]).transpose()
f.columns=['lead_id', 'order']
print(f.sort_values(by = 'order', ascending = False))
```
