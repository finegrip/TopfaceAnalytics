```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
from matplotlib import pyplot as plt
from sklearn.utils.extmath import cartesian

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
def active_next_week(df, sex, periods, is_alive_next_week, active_days_this_week, treshold = 4):
    for period_i in range(3,9):
        d = df[df['sex'] == sex]
        
        d = d[d['active_days' + periods[period_i]] == active_days_this_week]
        if is_alive_next_week:
            d = d[d['active_days' + periods[period_i+1]] >= treshold]
        else:
            d = d[d['active_days' + periods[period_i+1]] < treshold]
        d = d[[x for x in df.columns if periods[period_i] in x]]
        d = pd.concat( [d.agg(lambda x: np.sum(np.sign(x))), d.agg(lambda x: np.sum(np.sign(x))) / d.shape[0], d.mean(), d.quantile(0.25), d.quantile(0.5), d.quantile(0.75)], axis = 0)
        d.index = [i+'_'+j for j in ['sign','part','mean','quantile25','quantile50','quantile75'] for i in metrics]
        if('alive_metrix' in vars()):
            alive_metrix = pd.concat([alive_metrix, d], axis = 1)
        else:
            alive_metrix = d
    alive_metrix.columns = periods[3:9]
    return alive_metrix
```

```python
sex_dict = {0: 'women', 1: 'men'}
metrics = ["active_days","like_sent","fake_like_sent","like_received","fake_like_received","incoming_bomb","message_sent","message_received","mutual","fist_dialog_received","first_dialog_started","payment"]
periods = ["_0_1day","_1day_3day","_3day_7day","_7day_14day","_14day_21day","_21day_28day","_28day_35day","_35day_42day","_42day_49day","_49day_56day"]
# d = ["user_id","sex","isReferral","isPaid"] + [i+j for j in periods for i in metrics]
```

```python
treshold = 4

writer = pd.ExcelWriter('alive_metrix.xlsx')
for sex in [0,1]:
    if('alive_metrix' in vars()):
        del alive_metrix
    for period in periods:
        d = df[df['sex'] == sex]
        d = d[[x for x in df.columns if period in x]]
        d = d[d["active_days" + period] > 0]
        d = pd.concat( [d.agg(lambda x: np.sum(np.sign(x))), d.agg(lambda x: np.sum(np.sign(x))) / d.shape[0], d.mean(), d.quantile(0.25), d.quantile(0.5), d.quantile(0.75)], axis = 0)
        d.index = [i+'_'+j for j in ['sign','part','mean','quantile25','quantile50','quantile75'] for i in metrics]
        alive_metrix = pd.concat([alive_metrix, d], axis = 1) if 'alive_metrix' in vars() else d
    alive_metrix.columns = periods
    alive_metrix.to_excel(writer,'alive_metrix_' + sex_dict[sex])
    
    for active_days_this_week in range(5,8):
        alive_metrix = active_next_week(df = df, sex = sex, periods = periods, is_alive_next_week = True, active_days_this_week = active_days_this_week, treshold = treshold)
        nonalive_metrix = active_next_week(df = df, sex = sex, periods = periods, is_alive_next_week = False, active_days_this_week = active_days_this_week, treshold = treshold)
        # print(sex_dict[sex], active_days_this_week, alive_metrix.loc['like_sent_mean'])
        d = pd.concat([alive_metrix,nonalive_metrix], axis = 1)
        d.columns = [i + j for i in ['active','nonactive'] for j in alive_metrix.columns ]
        d = d[[i + j for j in alive_metrix.columns for i in ['active','nonactive']]]
        d.to_excel(writer,'act_next_week_' + str(active_days_this_week) + 'd_now_' + sex_dict[sex]) 
        
writer.save()
```
