`p.x.y`
p - Проценты 
x - Тип 0-3 
0 - Все время
1 - Ссесия
2 - Последние 30 симпатии
3 - Последние 10 сообщения

y - Метрики 0-7
0 - Процент положительного ответа на исходящие симпатии от пользователя (% ответа ему)
1 - Процент положительного ответа на симпатии
2 - Процент положительного ответа на исходящие взаимные симпатии от пользователя (% ответа ему)
3 - Процент положительного ответа на взаимные симпатии
4 - Процент ответа на первые сообщения
5 - Процент ответа на его первые сообщения
6 - Процент ответов на непервые сообщения
7 - Процент ответов на его непервые сообщения

`s.x.y`
s - Счетчики по которым считаются метрики.
x - Тип 0-3 
0 - Все время
1 - Ссесия
2 - Последние 30 симпатии
3 - Последние 10 сообщения

y - Счетчики 1-13
1 - Счетчик входящей активности, связанной с лайками
2 - Счетчики положительных ответов на исходящую симпатию
3 - Счетчики положительных ответов на входящую симпатию
4 - Счетчики ответов в процессе диалога, сколько ответил
5 - Счетчики отправленных и полученных взаимных симпатий
6 - Счетчики положительных ответов на исходящую взаимную симпатию
7 - Счетчики положительных ответов на входящую взаимную симпатию
8 - Счетчики ответов в процессе диалога, сколько ответили ему
9 - Счетчики первого контакта
10 - Счетчики ответов на первый контакт, сколько ответил
11 - Счетчики ответов на первый контакт, сколько ответили ему
12 - Счетчики ответов в процессе диалога, сколько ответил
13 - Счетчики ответов в процессе диалога, сколько ответили ему
