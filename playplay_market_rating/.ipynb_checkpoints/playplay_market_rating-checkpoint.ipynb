```python
import numpy as np
import pandas as pd
%matplotlib inline 
import re
from datetime import datetime
from datetime import timedelta
import chardet

pd.options.display.max_columns = 100
```

```python
# https://play.google.com/apps/publish/?account=5084250478732509234#BulkExportPlace:bep=com.topface.topface&bet=REVIEW&rwn=REVIEWS

with open('reviews%2Freviews_com.topface.topface_201803.csv', 'rb') as f:
    result1 = chardet.detect(f.read())  # or readline if the file is large
with open('reviews%2Freviews_com.topface.topface_201804.csv', 'rb') as f:
    result2 = chardet.detect(f.read())  # or readline if the file is large
with open('reviews%2Freviews_com.topface.topface_201805.csv', 'rb') as f:
    result3 = chardet.detect(f.read())  # or readline if the file is large

df = pd.concat([pd.read_csv('reviews%2Freviews_com.topface.topface_201803.csv', encoding=result1['encoding']),
                pd.read_csv('reviews%2Freviews_com.topface.topface_201804.csv', encoding=result2['encoding']),
                pd.read_csv('reviews%2Freviews_com.topface.topface_201805.csv', encoding=result3['encoding'])])
```

```python
df.head()
```

```python
df['date'] = df['Review Last Update Date and Time'].apply(lambda x: x[:10])
df['one'] = 1
```

```python
df.pivot_table(index='date', columns='Star Rating', values='one', aggfunc=len).fillna(0).applymap(float).to_csv('out.csv')
# df.groupby(['date','Star Rating']).sum()
```
